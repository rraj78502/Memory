```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np
```

```python
columns = ["age", "sex", "cp", "restbp", "chol", "fbs", "restecg",
"thalach", "exang", "oldpeak", "slope", "ca", "thal", "num"]
```

```python
df0 = pd.read_table("data/finalData.csv", sep=',', header=None,names=columns)
df0
```

```python
X = df0[df0.columns[1:13]]
X = np.array(X)
Y = df0[df0.columns[13:14]]
Y = np.array(Y)
```

```python
from keras.layers import Dense, Activation
model = Sequential()
#model.add(Dense(units=64, input_dim=(12,)))
#model.add(Activation('relu'))
#model.add(Dense(units=12))
#model.add(Activation('softmax'))
model.add(Dense(12, input_dim=12, init='uniform', activation='relu'))
model.add(Dense(4, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=2000, batch_size=16)
```

```python
scores = model.evaluate(X, Y)
print ("Accuracy: %.2f%%" %(scores[1]*100))
```

```python
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("model.h5")
print("Saved model to disk")
```

```python
import json
from keras.models import model_from_json
json_file = open('model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("model.h5")
print("Loaded model from disk")
# evaluate loaded model on test data
loaded_model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
score = loaded_model.evaluate(X, Y, verbose=0)
print("%s: %.2f%%" % (loaded_model.metrics_names[1], score[1]*100))
```
